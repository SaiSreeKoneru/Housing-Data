```python
import pandas as pd
df=pd.read_csv('ss13hil.csv',usecols=['HHL','HHT','WGTP','HINCP','ACCESS'], header=0,skip_blank_lines=True)
hhtdesc={
    1:'Married couple household',
 2:'Other family household:Male householder, no wife present',
 3:'Other family household:Female householder, no husband present', 
 4:'Nonfamily household:Male householder:Living alone',
 5:'Nonfamily household:Male householder:Not living alone', 
 6:'Nonfamily household:Female householder:Living alone',
 7:'Nonfamily household:Female householder:Not living alone'
}
hhldesc={
    1:'English only',
 2:'Spanish',
 3:'Other Indo-European languages',
 4:'Asian and Pacific Island languages',
 5:'Other language'
}
pd.set_option('display.max_column',500)
pd.set_option('display.max_colwidth',100)
pd.set_option('display.width',1000)
pd.set_option('display.float_format','{:.6f}'.format)
df['HHT'] = pd.to_numeric(df['HHT'], errors='coerce') 
df.dropna(subset=['HHT'], inplace=True)
print(df['HHT'].head(20))
df['HHT-assigned']=df['HHT'].apply(lambda id : hhtdesc[int(id)])
print(df['HHT-assigned'])
table1=df.groupby('HHT-assigned').agg({'HINCP':['mean','std','count','min','max']})
table1.columns=['mean','std','count','min','max']
table1.sort_values(by='mean',ascending=False,inplace=True)

table1['min']=table1['min'].astype(int)
table1['max']=table1['max'].astype(int)
print(table1)
```

```python
#tabel2
pd.set_option('display.max_column',500)
pd.set_option('display.max_colwidth',100)
pd.set_option('display.width',1000)
pd.set_option('display.float_format',"{:.2%}".format)
accdesc={1:'Yes w/ Subsrc.', 2:'Yes, wo/ Subsrc.', 3:'No'}

df['HHL'] = pd.to_numeric(df['HHL'], errors='coerce') 
df.dropna(subset=['HHL'], inplace=True)
df['ACCESS'] = pd.to_numeric(df['ACCESS'], errors='coerce') 
df.dropna(subset=['ACCESS'], inplace=True)
df['WGTP'] = pd.to_numeric(df['WGTP'], errors='coerce') 
df.dropna(subset=['WGTP'], inplace=True)

df['ACCESS']=df['ACCESS'].apply(lambda id: accdesc[int(id)])


df['HHL-assigned']=df['HHL'].apply(lambda id : hhldesc[int(id)])

table2=pd.crosstab(index=df['HHL-assigned'],values=df['WGTP'],aggfunc='sum',columns=df['ACCESS'],margins=True,normalize='all')
cust_columns=['Yes w/ Subsrc.','Yes, wo/ Subsrc.','No','All']
cust_index = ['English only', 'Asian and Pacific Island languages', 'Other Indo-European languages', 'Other language', 'Spanish', 'All']

table2 = table2.reindex(columns=cust_columns,index=cust_index)


print(table2) 


```

```python
#table3

pd.set_option('display.float_format',"{:.6f}".format)
df['HINCP'] = pd.to_numeric(df['HINCP'], errors='coerce') 
df.dropna(subset=['HINCP'], inplace=True)
grouping=pd.qcut(df['HINCP'],3,labels=['low','medium','high'])
grouped=df['HINCP'].groupby(grouping)
table3=grouped.describe()
display=['min','max','mean','count']
table3['min'] = table3['min'].round(0).astype(int)
table3['max'] = table3['max'].round(0).astype(int)
table3['count']=table3['count'].round(0).astype(int)

print(table3[display].rename(columns={'count':'household_count'}))
```
